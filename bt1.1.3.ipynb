```python
# Import th∆∞ vi·ªán
import pandas as pd
import numpy as np

# ƒê·ªçc d·ªØ li·ªáu
data = pd.read_csv("winequality-red.csv")
print("K√≠ch th∆∞·ªõc ban ƒë·∫ßu:", data.shape)
```

```python
# Ki·ªÉm tra s·ªë l∆∞·ª£ng d√≤ng tr√πng l·∫∑p
duplicates = data.duplicated().sum()
print(f"S·ªë d√≤ng tr√πng l·∫∑p: {duplicates}")

# Lo·∫°i b·ªè d√≤ng tr√πng l·∫∑p n·∫øu c√≥
data = data.drop_duplicates()
print("K√≠ch th∆∞·ªõc sau khi lo·∫°i b·ªè tr√πng l·∫∑p:", data.shape)
```

```python
# V√≠ d·ª•: ƒë·ªïi t√™n c·ªôt cho d·ªÖ ƒë·ªçc (n·∫øu c·∫ßn)
data.rename(columns={'volatile acidity': 'volatile_acidity',
                     'citric acid': 'citric_acid',
                     'residual sugar': 'residual_sugar',
                     'free sulfur dioxide': 'free_SO2',
                     'total sulfur dioxide': 'total_SO2'}, inplace=True)

# Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu c·ªßa c·ªôt 'quality' sang d·∫°ng category (ph√¢n lo·∫°i)
data['quality'] = data['quality'].astype('category')

# Ki·ªÉm tra l·∫°i ki·ªÉu d·ªØ li·ªáu
print("\nƒê·ªãnh d·∫°ng d·ªØ li·ªáu sau khi thay ƒë·ªïi:")
print(data.dtypes)
```

```python
# Ki·ªÉm tra d·ªØ li·ªáu thi·∫øu
missing = data.isnull().sum()
print("\nS·ªë l∆∞·ª£ng gi√° tr·ªã thi·∫øu:")
print(missing)
```

```python
# T√≠nh c√°c th·ªëng k√™ c∆° b·∫£n cho t·ª´ng c·ªôt (ch·ªâ l·∫•y c·ªôt s·ªë)
num_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Mean, Median, Mode
mean_vals = data[num_cols].mean()
median_vals = data[num_cols].median()
mode_vals = data[num_cols].mode().iloc[0]

# Variance, Standard deviation, Range
var_vals = data[num_cols].var()
std_vals = data[num_cols].std()
range_vals = data[num_cols].max() - data[num_cols].min()

# Percentiles v√† Quartiles
percentile_25 = data[num_cols].quantile(0.25)
percentile_50 = data[num_cols].quantile(0.5)
percentile_75 = data[num_cols].quantile(0.75)
iqr_vals = percentile_75 - percentile_25

# G·ªôp l·∫°i th√†nh m·ªôt b·∫£ng t·ªïng h·ª£p
desc_stats = pd.DataFrame({
    'Mean': mean_vals,
    'Median': median_vals,
    'Mode': mode_vals,
    'Variance': var_vals,
    'Std Dev': std_vals,
    'Range': range_vals,
    '25th Percentile': percentile_25,
    '50th Percentile': percentile_50,
    '75th Percentile': percentile_75,
    'IQR': iqr_vals
})

print("\nüìä B·∫£ng th·ªëng k√™ m√¥ t·∫£ chi ti·∫øt:")
print(desc_stats.round(3))
```
